```python
def snipper_auction(driver, players_type, action='bid', discount=0.8):
    listings = bidder.get_listings_info(driver)
    #print("Number of listings: ", len(listings))
    list_bid_prices=[]
    sucessfull_bids = 0
    failed_bids = 0
    found = 0
    for element in listings:
        
        try:
            auction_current_bid_int = get_current_bid(element)
            auction_initial_price = get_starting_price(element)
            rating = get_rating(element)
            name = utils.clean_characters(element.find_element_by_xpath('.//div[contains(@class, "name")]').text)
            auction_time = element.find_element_by_xpath('.//span[contains(@class, "time")]').text
            prebid_coins = utils.get_coins(driver)
            
#             print("Name: ", name, 
#               "Rating: ", rating,
#              "Starting Price: ",auction_initial_price, 
#              "Current BID: ", auction_current_bid_int, 
#              "Time Left: ", auction_time,
#              "Rating: ", rating)
        

        except:
            print("Error Retrieving Auction Data")
            pass
        
        try:
            # Compare with Futbin
            fut_price = max([auction_current_bid_int, auction_initial_price])
            futbin_price = compare_futbin(name, rating, fut_price, players_type)
            price_condition = (futbin_price*discount) > fut_price
            
        except:
            price_condition = False
            print("Error Comparing Prices")
            pass


        
        # Bid if conditions (Price and Time)
        if price_condition: #and auction_time == '<15 segundos':
            found = 1
            # Lowest default bid
            print("\n***********************")
            print("Make Bid on player: ", name, " Rating: ", str(rating))
            print("Time left: ", str(auction_time))
            print("Price FUT: ", fut_price)
            print("Futbin price: ", futbin_price)
            print("Max bid to make: ", futbin_price*0.9, " <----------")
            
            try:
                #print('---')
                if players_type == "Icon":
                    element.click()
                    driver.find_element_by_xpath('.//button[text()="Watch"]').click()
                else:
                    bidder.make_bid(element, driver)
                
            except:
                print("Error Bidding")
            
            #sleep(random.randint(5,10)/100)
            postbid_coins = utils.get_coins(driver)
            print("Pre Bid Coins: ", prebid_coins)
            print("Post Bid Coins: ", postbid_coins)
            print(postbid_coins-prebid_coins)
            os.system('say "Found player."')
            
            if postbid_coins < prebid_coins:
                
                print("Bid Made: ", prebid_coins - postbid_coins )
                sucessfull_bids += 1
                print("Succesfull BIDS ", sucessfull_bids)
            else:
                failed_bids += 1
                print("Failed Bid")
            print("*************************\n")
            
            if found == 1:
                break
            #sleep(random.randint(15,35)/10)
    
    return int(sucessfull_bids), int(failed_bids), found

#TODO Check players matching names
def compare_futbin(player, rating, price, players_type):
    # Get futbin data
    futbin_data = load_merge_futbin_catalogue()
    futbin_price = futbin_data.loc[(futbin_data.version==players_type) & 
                                   (futbin_data.name==player) &
                                   (futbin_data.rating==int(rating)), "price"]
    if (len(futbin_price) > 0) & (rating != ''):
        futbin_price = float(futbin_price.iloc[0])
        #print("---------", player, " Price in Futbin :", futbin_price, " Price in FUT: ", price, ' Max Price to BId: ', futbin_price*0.95)
        return futbin_price
    else:
        #print("Not Matched\n")
        return False

#TODO no leer en cada loop
def load_merge_futbin_catalogue():
    futbin_data = pd.read_csv('/Users/cognistx2019/Documents/GitHub/fifa21/data/FutBinCards21_2020_12_10_1309_all.csv')
    catalogue = pd.read_csv('/Users/cognistx2019/Documents/GitHub/fifa21/data/catalogos/new_catalogue.csv')
    futbin_data = futbin_data.merge(catalogue, on='full_name', how='inner').dropna(axis=0)
    return futbin_data
```

```python
import pandas as pd
import glob

path = r'/Users/cognistx2019/Documents/GitHub/fifa21/data' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

futbin_all = pd.concat(li, axis=0, ignore_index=True)
```

```python
futbin_all.date = pd.to_datetime(futbin_all.date)
futbin_all['day'] = futbin_all.date.dt.year.astype(str) + "-" + futbin_all.date.dt.month.astype(str) + "-" + futbin_all.date.dt.day.astype(str)
futbin_all = futbin_all.loc[futbin_all.price > 10, :]
futbin_all_grp = futbin_all.groupby(['full_name','rating','version','day'], group_keys=False).mean()
futbin_all_grp = futbin_all_grp.reset_index()
futbin_all_grp.day = pd.to_datetime(futbin_all_grp.day)
```

```python
def sort_growthrate(data):
    data = data.sort_values('day')
    data['gr'] = data.price.pct_change()
    data['ma_3'] = data.price.rolling(3).mean()
    return data
```

```python
data = futbin_all_grp.groupby(['full_name','rating','version'], group_keys=False).apply(lambda df : sort_growthrate(df))
```

```python
data.index = data.day
```

### Check who is going up

```python
# MA 3 days
ma_3 = data.loc[(data.price > data.ma_3)&(data.day == '2020-12-01')]
avg_gr = data.groupby(['full_name','rating','version'], 
                      group_keys=False).agg(mean_gr=('gr', 'mean'))


avg_gr = pd.DataFrame(avg_gr).reset_index()
positive_gr = avg_gr.loc[avg_gr.mean_gr > 0, :]
ma_gr = positive_gr.merge(ma_3, on=['full_name','rating','version'], how='inner')
ma_gr['ma3_dif'] = (ma_gr.price - ma_gr.ma_3)/ma_gr.ma_3
ma_gr.index = ma_gr.day
```
